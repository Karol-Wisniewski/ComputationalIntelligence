```python
import pandas as pd
import numpy as np
```

```python
iris_data: pd.DataFrame = pd.read_csv(
	"iris.csv",
	na_values={"NA", "-", "n/a", "na"},
	dtype={
		"sepallength": np.float64,
		"sepalwidth": np.float64,
		"petallength": np.float64,
		"petalwidth": np.float64,
		"class": pd.CategoricalDtype(categories=["setosa", "versicolor", "virginica"]),
	}
)
measurements_columns_names: list[str] = ["sepallength", "sepalwidth", "petallength", "petalwidth"]
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
standardized_iris_data: pd.DataFrame = iris_data.copy()
scaler: StandardScaler = StandardScaler()
for columnName in measurements_columns_names:
	standardized_iris_data[columnName] = scaler.fit_transform(iris_data[columnName].values.reshape(-1, 1))

standardized_iris_data
```
