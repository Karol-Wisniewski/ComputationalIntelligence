```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df = pd.read_csv('iris.csv')
```

Divide the data into training and test sets in the ratio 70:30

```python
X = df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']] # cechy
y = df['species'] # etykiety

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=274964)

# print("Train set:\n", X_train)
# print("Test set:\n", X_test)
```

```python
clf = DecisionTreeClassifier()
```

Training the decision tree classifier

```python
clf.fit(X_train, y_train)
```

Text representation of the decision tree

```python
tree.plot_tree(clf)
```

Evaluation of the classifier

```python
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

The elements on the main diagonal of the matrix represent correct predictions (i.e., the number of samples that actually belong to a given class and were correctly classified). The elements off the main diagonal represent incorrect predictions (i.e., the number of samples that actually belong to one class but were mistakenly classified as another class).

```python
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", cm)
```
